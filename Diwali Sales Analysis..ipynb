****State****

```python
sales_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by=['Orders'], ascending=False)
```

```python
sales_state
```

```python
#Top 10 states by orders
sales_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by=['Orders'], ascending=False).head(10)
#(.head) function should apply at end

sns.set(rc={'figure.figsize':(15,5)}) #(width,height)

#RC parameter is used to manage the size of the figure. We are passing the value of the dictionary (key/value pairs) to the parameter by
#using the key as {'figure.figsize':(value)} and the dimensions required

sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
#Top 10 states by amount
sales_state = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by=['Amount'], ascending=False).head(10)
#(.head) function should apply at end

sns.set(rc={'figure.figsize':(15,5)}) #(width,height)

#RC parameter is used to manage the size of the figure. We are passing the value of the dictionary (key/value pairs) to the parameter by
#using the key as {'figure.figsize':(value)} and the dimensions required

sns.barplot(data = sales_state, x = 'State', y = 'Amount')
```

***from above graphs we can see that most of the order ar from Uttar Pradesh, Maharashtra and Karnataka respectively***  

**MARITAL STATUS**

```python
df.columns
```

```python
#shopping vs marital status
svms = sns.countplot(data = df, x= 'Marital_Status')
sns.set(rc={'figure.figsize':(1,5)})

for bars in svms.containers:
    svms.bar_label(bars)
```

```python
#Marital status, gender vs amount
smga = df.groupby(['Marital_Status', 'Gender'], as_index= False)['Amount'].sum().sort_values(by=['Amount'], ascending=False).head(10)

sns.set(rc={'figure.figsize':(5,5)})

sns.barplot(data = smga, x = 'Marital_Status',  y = 'Amount',hue = 'Gender')
```

**OCCUPATION**

```python
sns.set(rc={'figure.figsize':(20,5)})
OA = sns.countplot(data = df, x = 'Occupation')
for bars in OA.containers:
    OA.bar_label(bars)
```

```python
#sort by count
sns.set(rc={'figure.figsize':(20,5)})
OA = sns.countplot(data = df, x = 'Occupation', order=df['Occupation'].value_counts().index)
for bars in OA.containers:
    OA.bar_label(bars)
```

```python
#occupation va amount
OA = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
```

```python
OA
```

```python
#occupation va amount
OA = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
OA = sns.barplot(data = OA, x = 'Occupation', y = 'Amount')

for bars in OA.containers:
    OA.bar_label(bars)
```

***from above graph we can see that most of the buyers are from IT Sector, Healthcare and Aviation***

***Product Category***

```python
sns.set(rc={'figure.figsize':(30,5)})
PCA = sns.countplot(data = df, x = 'Product_Category',order=df['Product_Category'].value_counts().head(10).index)
for bars in PCA.containers:
    PCA.bar_label(bars)
```

```python
#product and Amount
PCA = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by= ['Amount'], ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
PCA = sns.barplot(data = PCA, x = 'Product_Category', y = 'Amount')
for bars in PCA.containers:
    PCA.bar_label(bars)
```

***from the above graph we can see that most sold product are food, clothing and electronic category***

***ProductID and order***

```python
#ProductID and order
PDO = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by= ['Orders'], ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
PDO = sns.barplot(data = PDO, x = 'Product_ID', y = 'Orders')

for bars in PDO.containers:
    PDO.bar_label(bars)
```

***Conclusion***

***Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, healthcare, and Aviation are more likely buy products from food ,clothing and electronics category***
